```python
import pandas as pd

url = "https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Sales_Data/Sales_April_2019.csv"
df = pd.read_csv(url)

df.head()
```

```python
# Clean data properly
df = df.dropna()
df = df[df['Order Date'] != 'Order Date']

# Convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Drop again (important 🔥)
df = df.dropna()

# Extract month
df['Month'] = df['Order Date'].dt.month
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])

df['Sales'] = df['Quantity Ordered'] * df['Price Each']
```

```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby('Month')['Sales'].sum()

plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Analysis")
plt.show()
```

```python
top_products = df.groupby('Product')['Quantity Ordered'].sum()
top_products.sort_values(ascending=False).head()
```
